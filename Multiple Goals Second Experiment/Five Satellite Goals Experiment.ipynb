```python
import numpy as np
from sgp4.api import Satrec
from poliastro.twobody import Orbit
from poliastro.plotting import OrbitPlotter3D, OrbitPlotter2D
from poliastro.bodies import Earth
from astropy import units as u
```

```python
class Satellite:
    
    dv_limit = 1*u.km/u.s #km/s
    
    def __init__(self, s, t):
        self.s = s
        self.t = t
        self.satellite = Satrec.twoline2rv(s,t)
        
    def getVec(self,jd,fr):
        return ([i for i in self.satellite.sgp4(jd,fr)[1]]*u.km,[i for i in self.satellite.sgp4(jd,fr)[2]]*u.km/u.s)
```
